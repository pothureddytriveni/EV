######1. load the iris.csv dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("Iris.csv")
df=pd.DataFrame(df)
print(df)
```

#### 1. create a bar plot for count the frequency of the three species.

```python
ax=plt.subplots(1,1,figsize=(6,4))
sns.countplot('Species',data=df)
plt.title("Iris Species Count")
plt.show()
```

##### 2. Create a pie chart to show the percentage of species.

```python
df['Species'].value_counts().plot.pie(autopct='%1.1f%%',figsize=(6,4))
plt.title("percentage of species")
plt.show()
```

##### 3. Create the joint to all numberical variable for analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
sns.jointplot(df.SepalLengthCm,df.SepalWidthCm,kind="kde")
```

##### 4. Create a scatter plot and take hue = "Species"

```python
sns.scatterplot(data=df,x="SepalLengthCm",y="PetalLengthCm",hue="Species")
```

###### 5. Create a Boxplot or Whisker plot as univeriate and bivariate analysis.

```python
sns.boxplot(df["SepalLengthCm"])
plt.show()
```

```python
sns.boxplot(df["SepalLengthCm"],df["PetalLengthCm"])

plt.show()
```

###### 6. Create a Strip plot 

```python
sns.stripplot(df["PetalLengthCm"])
plt.show()
```

###### 7. Create a Violin Plot.

```python
sns.violinplot(df["PetalLengthCm"])
```

##### 8. Create a pair chart.

```python
sns.pairplot(data = df, vars = ["SepalLengthCm", "PetalLengthCm"])
plt.show()
```

##### 9. Create a Heat Map

```python
sns.heatmap(df[["SepalLengthCm","PetalLengthCm"]].corr(),annot=True)
```

###### 10. Create a distribution plot for all numerical colomns.

```python
sns.distplot(df[["SepalLengthCm", 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']])
plt.show()
```
